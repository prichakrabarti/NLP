```python
#Convert to lowercase
df["Phrase"]= df["Phrase"].apply(lambda x:x.lower())
```

```python
df["Phrase"]= df["Phrase"].apply(lambda x: word_tokenize(x))
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = set(stopwords.words('english'))
```

```python
len(stopwords)
```

```python
df['Phrase']= df['Phrase'].apply(lambda x: [item for item in x if item not in stopwords])
```

```python
X= df["Phrase"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf= TfidfVectorizer()
```

```python
X= tfidf.fit_transform(df["Phrase"])
```

```python
print(tfidf.vocabulary_)
```

```python
y= df["Sentiment"]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X,y, test_size= 0.3, random_state= 42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb= MultinomialNB()
```

```python
mnb.fit(X_train, y_train)
```

```python
y_pred= mnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score (y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, y_pred)
```

```python
#Lets try an SVM model
from sklearn.svm import SVC
```

```python
svc= SVC(kernel = "linear")
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred_svc = svc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_svc)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg= LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
y_pred_logreg= logreg.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_logreg)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier()
```

```python
ada.fit(X_train, y_train)
```

```python
y_pred_ada = ada.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_ada)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb= GaussianNB()
```

```python
gnb.fit(X_train.toarray(), y_train)
```

```python
y_pred_gnb= gnb.predict(X_test.toarray())
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_gnb)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred_knn= knn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_knn)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators=[('knn', knn), ('logreg', logreg), ('SVM', svc)])
```

```python
vc.fit(X_train, y_train)
```

```python
y_pred_vc= vc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_vc)
```
